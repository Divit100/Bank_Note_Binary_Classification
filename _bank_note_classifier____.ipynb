```python
import numpy as np
```

```python
from numpy import genfromtxt
data=genfromtxt('C:/Users/panindia/Desktop/Udemy/Computer-Vision-with-Python/DATA/bank_note_data.txt',delimiter=',')
```

```python
features=data[:,0:4]
labels=data[:,4]
```

```python
X=features 
y=labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler_object=MinMaxScaler()
```

```python
scaler_object.fit(X_train)
```

```python
scaler_X_train=scaler_object.transform(X_train)
scaler_X_test=scaler_object.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense 

model=Sequential()

model.add(Dense(4,input_dim=4,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(scaler_X_train,y_train,epochs=50,verbose=2)
```

```python
model.predict_classes(scaler_X_test)
```

```python
model.metrics_names
```

```python
model.evaluate(x=scaler_X_test,y=y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report 
```

```python
predictions=model.predict_classes(scaler_X_test)
```

```python
confusion_matrix(y_test,predictions )
```

```python
print(classification_report (y_test,predictions))
```
